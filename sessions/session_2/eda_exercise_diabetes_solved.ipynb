### Exercise 4: Handling Missing Data (Implicit Missing Values)

As observed in Exercise 1.3, some columns have a minimum value of 0, which might represent missing data (e.g., a blood pressure of 0 is biologically impossible).

1. For the `Glucose`, `BloodPressure`, `SkinThickness`, `Insulin`, and `BMI` columns, replace the 0 values with `NaN` (Not a Number) to explicitly mark them as missing.
2. Verify the number of missing values in these columns after the replacement using `.info()` or `.isnull().sum()`. Which column has the most missing values now?
3. Choose an imputation strategy for the missing numerical values (e.g., replace with the mean or median of the respective column). Justify your choice.
4. Apply the chosen imputation strategy to fill the `NaN` values in the affected columns. Verify that there are no more missing values in these columns.

```python
# Your code for Exercise 4.1
cols_with_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
diabetes_df[cols_with_zeros] = diabetes_df[cols_with_zeros].replace(0, np.nan)
```

```python
# Your code for Exercise 4.2
missing_values_after_replace = diabetes_df.isnull().sum()
print("Missing values after replacing 0s:")
print(missing_values_after_replace)

most_missing_col = missing_values_after_replace.idxmax()
print(f"Column with the most missing values: {most_missing_col}")
```

```python
# Your code for Exercise 4.3 (Add comments for justification)
# Justification: Using the median is often preferred over the mean for imputation when the data might have outliers, as the median is less affected by extreme values.
imputation_strategy = diabetes_df[cols_with_zeros].median()
```

```python
# Your code for Exercise 4.4
diabetes_df[cols_with_zeros] = diabetes_df[cols_with_zeros].fillna(imputation_strategy)

# Verify no more missing values in these columns
print("Missing values after imputation:")
print(diabetes_df[cols_with_zeros].isnull().sum())
```

### Exercise 5: Creating New Features (Feature Engineering)

Create a new categorical feature based on the `Age` column.

1. Define age bins (e.g., '0-20', '21-30', '31-40', '41-50', '51+').
2. Create a new column called `AgeGroup` by categorizing the `Age` values into these bins.
3. Display the value counts for the new `AgeGroup` column.

```python
# Your code for Exercise 5.1, 5.2, and 5.3
bins = [0, 20, 30, 40, 50, np.inf]
labels = ['0-20', '21-30', '31-40', '41-50', '51+']
diabetes_df['AgeGroup'] = pd.cut(diabetes_df['Age'], bins=bins, labels=labels, right=True)

print("Value counts for AgeGroup:")
print(diabetes_df['AgeGroup'].value_counts())
```

```python
# Example for Option 3: Capping outliers in Insulin (using 95th percentile as an example cap)
# Note: This is just one simple method, and the choice of percentile can be debated.
cap_value = diabetes_df['Insulin'].quantile(0.95)
diabetes_df['Insulin_capped'] = diabetes_df['Insulin'].clip(upper=cap_value)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Insulin', data=diabetes_df)
plt.title('Boxplot of Insulin Before Capping')
plt.show()
plt.figure(figsize=(8, 6))
sns.boxplot(x='Insulin_capped', data=diabetes_df)
plt.title('Boxplot of Insulin After Capping at 95th Percentile')
plt.xlabel('Insulin (Capped)')
plt.show()
```
